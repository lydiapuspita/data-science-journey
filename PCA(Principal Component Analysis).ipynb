## PCA
### Manual Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean()
```

```python
scaled_X - scaled_X.mean(axis=0)
```

```python
scaled_X
```

```python
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_vectors.shape
```

```python
# PCA N Feature Space ---> N PC Space
# How many PC to choose --> N = 2
# PCA 30 Feature Space ---> 30 PC Space ---> 2
num_components = 2
```

```python
# [0,1,2]
sorted_key = np.argsort(eigen_values)[::-1][num_components]
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

```python
# Original data --> Project --> eigen_vectors
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
principal_components
```

```python
plt.figure(figsize=(8,6),dpi=150)
plt.scatter(principal_components[:,0],principal_components[:,1])
```

### Scikit Learn

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
# decomposition from a large space to a smaller one
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
# there are 2 ways to do this
pca_model.fit(scaled_X)
pca_model.transform(scaled_X)
```

```python
# or
pc_result = pca_model.fit_transform(scaled_X)
```

```python
plt.figure(figsize=(8,6),dpi=150)
plt.scatter(pc_result[:,0],pc_result[:,1],c=cancer_dictionary['target'])
```

```python
cancer_dictionary['target']
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_,index=['PC1','PC2'],columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20,3),dpi=150)
sns.heatmap(df_comp,annot=True)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
# the total variance ratio should be 100%, so we must fit in the pca_model on the better n_component
# let's try n_component=30
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_) #--> this is what should happen
```

```python
# or we can use for loop to find out the best result of n_components
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel('Num of Components')
plt.ylabel('Variance Explained')
```
